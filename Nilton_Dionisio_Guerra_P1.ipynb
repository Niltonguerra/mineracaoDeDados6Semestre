```python
# Criar bins para agrupar os preços e as pontuações
price_bins = np.logspace(0.5, 4, 50)  # Bins para preços (logaritmicamente espaçados)
points_bins = np.linspace(df1['points'].min(), df1['points'].max(), 50)  # Bins para pontuação

# Criar uma tabela de frequência (2D histogram) usando np.histogram2d
heatmap_data, xedges, yedges = np.histogram2d(df1['price'], df1['points'], bins=[price_bins, points_bins])

# Plotar o Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data.T, cmap="Purples", xticklabels=np.round(xedges, 1), yticklabels=np.round(yedges, 1))

# Ajustar eixos e título
plt.title('Relação entre Preço e Pontuação (Heatmap)', fontsize=14)
plt.xlabel('Preço ($)')
plt.ylabel('Pontuação')
plt.show()
```

```python
import re
```

```python
# Função para extrair datas
def extrair_data(nome):
    # Regex para encontrar anos no formato de 4 dígitos
    match = re.search(r'\b\d{4}\b', nome)
    return match.group(0) if match else None

df2 = df1.copy()

df2 = df2.drop(["description","province","variety","winery"], axis=1)

# Aplicando a função ao DataFrame
df2['data_de_criacao'] = df1['title'].apply(extrair_data)

print(df2)
```

```python
#converte para numero
df2['data_de_criacao'] = pd.to_numeric(df2['data_de_criacao'], errors='coerce')

# Calcular a idade do vinho
ano_atual = 2024  # Substitua pelo ano atual se necessário

df2 = df2[(df2['data_de_criacao'] > 1900) & (df2['data_de_criacao'] <= ano_atual)]

# Calcular a idade do vinho (ano atual - data_de_criacao)
df2['idade'] = ano_atual - df2['data_de_criacao']

# Filtrar idades negativas ou valores muito grandes
df2 = df2[(df2['idade'] >= 0) & (df2['idade'] <= 100)]

# Remover linhas onde 'idade' ou 'price' é NaN (dados faltantes)
df2.dropna(subset=['idade', 'price'], inplace=True)

```

```python

# Configurar o tamanho da figura
plt.figure(figsize=(12, 8))

sns.regplot(x='idade', y='price', data=df2, scatter_kws={'alpha':0.5}, color='Purple')

# Adicionar título e rótulos
plt.title('Relação entre Idade do Vinho e Preço', fontsize=16)
plt.xlabel('Idade do Vinho (anos)', fontsize=14)
plt.ylabel('Preço ($)', fontsize=14)

# Exibir o gráfico
plt.show()
```

```python
# Agrupar por país e calcular o preço médio
media_preco_pais = df1.groupby('country')['price'].mean()

# Ordenar em ordem decrescente
media_preco_pais_ordenado = media_preco_pais.sort_values(ascending=False)

# Exibir os 10 países com vinhos mais caros
top_paises = media_preco_pais_ordenado.head(10)
print(top_paises)

# Plotar os resultados
plt.figure(figsize=(8,6))
sns.barplot(x=top_paises.values, y=top_paises.index, palette='Purples')
plt.title('Top 10 Países com Vinhos Mais Caros', fontsize=16)
plt.xlabel('Preço Médio ($)', fontsize=14)
plt.ylabel('País', fontsize=14)
plt.show()
```
