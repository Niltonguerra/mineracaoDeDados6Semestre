```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('/content/winemag-data-130k-v2.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["country"].value_counts()
```

```python
df["variety"].value_counts()
```

```python
df1 = df.copy()
```

```python
df1 = df1.drop(["Unnamed: 0", "designation", "region_1", "region_2", "taster_name", "taster_twitter_handle"], axis=1)
df1.head()
```

```python
df1.info()
```

```python
# verifica o tamanho da tabela
df1.shape
```

```python
#verifica o tamanho da tabela se fosse removidos todos os dados com dados nulos
print("{:.0%}".format(df1["price"].count() / df1.shape[0]))
```

```python
df1 = df1.dropna(subset=["country","price","variety"])
df1.info()
```

```python
# análise exploratória
plt.figure(figsize=(10,7))
plt.hist(data=df1, x='points',
         bins=np.arange (80,df1['points'].max()+1,1),
         color= 'purple')
plt.xlim(80,100)
plt.title('Distribuição de Pontuação', fontsize= 16)
plt.xlabel('Pontuação')
plt.ylabel('Quantidade')
```

```python
country_top_10 = df1['country'].value_counts()[:10]
country_top_10
```

```python
plt.figure(figsize=(10,7))
country_top_10.plot(kind = 'bar', color = 'purple')
plt.title('Número de classificações de vinhos por país', fontsize= 16)
plt.xlabel('País')
plt.ylabel('Total de avaliações')
```

```python
melhor_qualidade = df1.groupby ("country")['points'].agg('mean')
melhor_qualidade = melhor_qualidade.sort_values(ascending = False)[:10]
#converter um uma série o data frame usando o método to_frame()
melhor_qualidade.to_frame().style.background_gradient(cmap='Purples', high=0.5)
```

```python
high_scrores = df1[df1['points'] >= 95]
high_scrores["country"].value_counts()
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(data=df1, x= 'points',y='country',color='Purple')
plt.title('Distribuição de Pontuação por País', fontsize= 16)
plt.xlabel('Pontuação')
plt.ylabel('País')

```

```python
average_price = df1.groupby ("country")['price'].mean()
average_price = average_price.sort_values(ascending = False)
y = average_price.index.to_series()
x = average_price.values

plt.figure(figsize=(12,8))
sns.barplot(x=x, y=y, palette='Purples_r')
plt.title('Preço médio dos vinhos por País', fontsize= 16)
plt.xlabel('Preço Médio')
plt.ylabel('País')

```

```python
#verifica a distribuição dos preços dos vinhos
plt.figure(figsize=(10,7))
bins = np.arange (0, df1['price'].max()+5, 5)
plt.hist(data=df1,x='price',bins= bins, color= 'Purple')
plt.xlim(0,200)
plt.title('Distribuição dos preços dos vinhos', fontsize= 16)
plt.xlabel('preço($)')
plt.ylabel('Frequência')
```

```python
df1['price'].describe()
```

```python
np.log10(df1['price'].describe())
```

```python
plt.figure(figsize=(12,8))
bins = 10 ** np.arange(0.5,3.5 + 0.05,0.05)
ticks = [1,3,10,30,100,300,1000,3000,10000]
plt.hist(data=df1, x='price', bins=bins, color='Purple')
plt.xscale('log')
plt.xticks(ticks,ticks)
plt.xlabel('preço($)')
plt.ylabel('Frequência')
```

```python
x = [1,3,10,30,100,300,1000,3000,10000]
plt.figure(figsize = (12,8))
sns.regplot(x='price',y='points',data=df1, color = 'Purple', logx=True, scatter_kws={'alpha':0.3})
plt.title('Relação entre preço e pontuação', fontsize= 14)
plt.xscale('log')
plt.xticks(x,x)
plt.xlabel('preço($)')
plt.ylabel('Pontuação')
```

```python
# Criar bins para agrupar os preços e as pontuações
price_bins = np.logspace(0.5, 4, 50)  # Bins para preços (logaritmicamente espaçados)
points_bins = np.linspace(df1['points'].min(), df1['points'].max(), 50)  # Bins para pontuação

# Criar uma tabela de frequência (2D histogram) usando np.histogram2d
heatmap_data, xedges, yedges = np.histogram2d(df1['price'], df1['points'], bins=[price_bins, points_bins])

# Plotar o Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data.T, cmap="Purples", xticklabels=np.round(xedges, 1), yticklabels=np.round(yedges, 1))

# Ajustar eixos e título
plt.title('Relação entre Preço e Pontuação (Heatmap)', fontsize=14)
plt.xlabel('Preço ($)')
plt.ylabel('Pontuação')
plt.show()
```
