# 01 - Criação e Manipulação de Arrays

Crie um array Numpy com 10 números inteiros aleatórios entre 1 e 20 e extraia os elementos pares.

```python
import numpy as np

# Cria um array Numpy com 10 números inteiros aleatórios entre 1 e 20
array = np.random.randint(1, 21, size=10)

# Extrai os elementos pares do array
pares = array[array % 2 == 0]

# Exibe o array original e os elementos pares
print("Array original:", array)
print("Elementos pares:", pares)
```

# 02 - Operações Aritméticas com Arrays

Crie dois arrays de 10 números inteiros aleatórios entre 0 e 50 e realize operações de soma, subtração, multiplicação e divisão elemento a elemento.

```python
import numpy as np

# Cria dois arrays Numpy com 10 números inteiros aleatórios entre 0 e 50
array1 = np.random.randint(0, 51, size=10)
array2 = np.random.randint(0, 51, size=10)

# Realiza operações elemento a elemento
soma = array1 + array2
subtracao = array1 - array2
multiplicacao = array1 * array2
# Para evitar divisão por zero, substitui zeros em array2 por 1 temporariamente
divisao = np.divide(array1, np.where(array2 == 0, 1, array2))

# Exibe os arrays originais e os resultados das operações
print("Array 1:", array1)
print("Array 2:", array2)
print("Soma:", soma)
print("Subtração:", subtracao)
print("Multiplicação:", multiplicacao)
print("Divisão:", divisao)
```

# 03 - Indexação e Slicing

Dado um array Numpy com 20 números aleatórios entre 10 e 100, extraia os elementos entre a 5ª e a 15ª posição.

# 04 - Manipulação de Matrizes

Crie uma matriz 3x3 com valores aleatórios entre 0 e 1, e encontre o valor máximo e mínimo.

```python
import numpy as np

# Cria uma matriz 3x3 com valores aleatórios entre 0 e 1
matriz = np.random.rand(3, 3)

# Encontra o valor máximo e mínimo na matriz
valor_maximo = np.max(matriz)
valor_minimo = np.min(matriz)

# Exibe a matriz e os valores máximo e mínimo
print("Matriz 3x3 com valores aleatórios:")
print(matriz)
print("Valor máximo:", valor_maximo)
print("Valor mínimo:", valor_minimo)
```

# 06 - Criação de DataFrames

Crie um DataFrame com dados de 5 pessoas, onde cada pessoa tem nome, idade (entre 20 e 40), e cidade (escolha 3 cidades aleatórias).

```python
import pandas as pd
import numpy as np

# Nomes das pessoas
nomes = ['Ana', 'Bruno', 'Carla', 'Diego', 'Elena']

# Idades aleatórias entre 20 e 40
idades = np.random.randint(20, 41, size=5)

# Cidades aleatórias escolhidas de uma lista de 3 cidades
cidades_possiveis = ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte']
cidades = np.random.choice(cidades_possiveis, size=5)

# Criação do DataFrame
dados = {'Nome': nomes, 'Idade': idades, 'Cidade': cidades}
df = pd.DataFrame(dados)

# Exibe o DataFrame
print(df)
```

# 07 - Criação de DataFrame com Dados Sintéticos

Crie um DataFrame fictício de vendas contendo colunas como Produto, Quantidade (entre 1 e 20), e Preço (entre 10 e 100).

```python
import pandas as pd
import numpy as np

# Lista de produtos fictícios
produtos = ['Produto A', 'Produto A', 'Produto C', 'Produto D', 'Produto E']

# Quantidade aleatória entre 1 e 20 para cada produto
quantidades = np.random.randint(1, 21, size=5)

# Preço aleatório entre 10 e 100 para cada produto
precos = np.random.randint(10, 101, size=5)

# Criação do DataFrame
dados_vendas = {'Produto': produtos, 'Quantidade': quantidades, 'Preço': precos}
df_vendas = pd.DataFrame(dados_vendas)

# Exibe o DataFrame
print(df_vendas)
```

# 08 - Filtragem de Dados

Filtre o DataFrame criado no exercício anterior para mostrar apenas as vendas onde a quantidade é maior que 10.

```python
# Filtra o DataFrame para mostrar apenas as vendas onde a quantidade é maior que 10
vendas_filtradas = df_vendas[df_vendas['Quantidade'] > 10]

# Exibe o DataFrame filtrado
print(vendas_filtradas)
```

# 09 - Operações em Colunas

Adicione uma coluna ao DataFrame com o valor total da venda (Quantidade * Preço).

```python
# Adiciona uma coluna 'Valor Total' ao DataFrame com o valor total da venda (Quantidade * Preço)
df_vendas['Valor Total'] = df_vendas['Quantidade'] * df_vendas['Preço']

# Exibe o DataFrame atualizado
print(df_vendas)
```

# 10 - Agrupamento e Agregação

Agrupe o DataFrame pelo campo Produto e calcule a soma das quantidades vendidas para cada produto.

```python
# Agrupa o DataFrame pelo campo 'Produto' e calcula a soma das quantidades vendidas para cada produto
quantidade_total_por_produto = df_vendas.groupby('Produto')['Quantidade'].sum().reset_index()

# Exibe o DataFrame com a soma das quantidades
print(quantidade_total_por_produto)
```

# 11 - Gráfico de Linhas

Crie um gráfico de linha que mostra a variação de preços de um produto fictício ao longo de 12 meses.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Criando dados fictícios
meses = np.arange(1, 13)  # Meses de 1 a 12
precos = np.random.uniform(10, 100, size=12)  # Preços aleatórios entre 10 e 100

# Criando um DataFrame com os dados
df_precos = pd.DataFrame({
    'Mês': meses,
    'Preço': precos
})

# Criando o gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(df_precos['Mês'], df_precos['Preço'], marker='o', linestyle='-', color='b')
plt.title('Variação de Preços ao Longo de 12 Meses')
plt.xlabel('Mês')
plt.ylabel('Preço')
plt.xticks(ticks=meses)  # Define os ticks no eixo x para mostrar todos os meses
plt.grid(True)
plt.show()
```

# 17 - Gráfico de Dispersão com Seaborn

Recrie o gráfico de dispersão da relação entre altura e peso usando Seaborn, incluindo uma linha de regressão.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Gerar dados fictícios
np.random.seed(0)  # Para garantir a reprodutibilidade
altura = np.random.uniform(150, 200, size=100)  # Alturas entre 150 e 200 cm
peso = np.random.uniform(50, 100, size=100)     # Pesos entre 50 e 100 kg

# Criando um DataFrame com os dados
df = pd.DataFrame({
    'Altura': altura,
    'Peso': peso
})

# Criando o gráfico de dispersão com linha de regressão
plt.figure(figsize=(10, 6))
sns.regplot(x='Altura', y='Peso', data=df, scatter_kws={'color':'coral', 's':50}, line_kws={'color':'blue'})

# Adicionando título e rótulos de eixo
plt.title('Relação entre Altura e Peso com Linha de Regressão')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.grid(True, linestyle='--', alpha=0.7)

# Exibindo o gráfico
plt.show()
```

# 19 - Heatmap

Crie um heatmap para mostrar a correlação entre diferentes variáveis em um DataFrame fictício.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Gerar dados fictícios
np.random.seed(0)  # Para garantir a reprodutibilidade
data = {
    'Variável A': np.random.randn(100),  # Dados normalmente distribuídos
    'Variável B': np.random.randn(100),
    'Variável C': np.random.randn(100),
    'Variável D': np.random.randn(100)
}

# Criando um DataFrame com os dados
df = pd.DataFrame(data)

# Calculando a matriz de correlação
correlation_matrix = df.corr()

# Criando o heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Adicionando título
plt.title('Matriz de Correlação entre Variáveis')

# Exibindo o gráfico
plt.show()
```

# 20 - Pairplot

Crie um pairplot para explorar relações entre várias variáveis em um conjunto de dados fictício.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Gerar dados fictícios
np.random.seed(0)  # Para garantir a reprodutibilidade
data = {
    'Variável A': np.random.randn(100),  # Dados normalmente distribuídos
    'Variável B': np.random.randn(100),
    'Variável C': np.random.randn(100),
    'Variável D': np.random.randn(100)
}

# Criando um DataFrame com os dados
df = pd.DataFrame(data)

# Criando o pairplot
sns.pairplot(df, diag_kind='hist', plot_kws={'alpha':0.6, 'edgecolor': 'k'}, diag_kws={'color':'blue'})

# Adicionando título
plt.suptitle('Pairplot das Variáveis', y=1.02)

# Exibindo o gráfico
plt.show()

```
